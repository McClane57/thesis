```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 0.57*(x/10.-10)**2

def grad(x):
    return 0.57/5.*(x/10.-10)

def fgrad(x):
    return f(120) - grad(120)*(120-x)

def ff(x, y, L):
    return 0.57*(y/10.-10)**2 + L/2.*(x-y)**2 - grad(120)*(120-x
                                                          )
```

```python
xx = np.arange(300)
```

```python
yy = f(xx)
```

```python
zz2 = ff(xx, 120, 0.02)
zz = ff(xx, 120, 0.0057)
```

```python
gg = fgrad(xx)
```

```python
plt.plot(xx,yy, color = 'black')
plt.ylim(-10,100)
plt.plot(xx,zz, ':', color = 'blue')
plt.plot(xx,zz2, ':', color = 'green')
plt.plot(xx,gg, 'r')
plt.scatter(120, f(120), color = 'black')
plt.text(125,0, "$text$", color = 'black')
plt.text(225, f(120)+10, "$(x, f(x))$", color = 'black')

plt.text(0, f(120)+10, "$(x, f(x))$", color = 'black')
plt.text(0, f(120)+10, "$(x, f(x))$", color = 'black')
plt.text(0, f(120)+10, "$(x, f(x))$", color = 'black')
#plt.axis('off')
```
